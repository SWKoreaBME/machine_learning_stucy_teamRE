```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import normalized_mutual_info_score, adjusted_mutual_info_score
from sklearn.feature_selection import mutual_info_classif, SelectKBest, chi2

import random

'''
    Dataset : Heart Disease UCI https://www.kaggle.com/ronitf/heart-disease-uci/version/1
    Target : Presence of heart disease in the patient
    
'''

data_file = './heart.csv'

heart_df = pd.read_csv(data_file)

cols = list(heart_df.columns[:-1])

whole_data = [data.values for i, data in heart_df.iterrows()]

random.shuffle(whole_data)

X = [data[:-1] for data in whole_data] # each row is in generator
Y = [data[-1] for data in whole_data]
```

## Pearson's correlation Coefficient

1. calculate some Pearson's Correlation Coefficient between each features

```python
'''
    Pearson's Correlation Coefficient
'''

# age = [data[0] for data in X]
# sex = [data[1] for data in X]

correlation_matrix = np.corrcoef(np.transpose(X))
```

```python
sns.set()

f, ax = plt.subplots(figsize = (9,6))

ax = sns.heatmap(correlation_matrix, annot=True)
ax.set_title('Correlation Coefficient')
```

## Mutual information score

1. calculate mutual information score for each feature distributions
2. 1 ( perfect correlation ) , 0 ( no mutual information )

```python
features = np.transpose(X)

MI_matrix = np.ones((13,13))

for i in range(13):
    
    for j in range(i+1, 13):
        
        MI_score = normalized_mutual_info_score(features[i], features[j])
        
        MI_matrix[i][j] = MI_score
        MI_matrix[j][i] = MI_score
```

```python
sns.set()

f, ax = plt.subplots(figsize = (9,6))

ax = sns.heatmap(MI_matrix, annot=True)
ax.set_title('Normalized Mutual information Score')
```


### Bonus : Adjusted Mutual information
    
1. https://scikit-learn.org/stable/modules/generated/sklearn.metrics.adjusted_mutual_info_score.html#sklearn.metrics.adjusted_mutual_info_score


```python
features = np.transpose(X)

AMI_matrix = np.ones((13,13))

for i in range(13):
    
    for j in range(i+1, 13):
        
        MI_score = adjusted_mutual_info_score(features[i], features[j])
        
        AMI_matrix[i][j] = MI_score
        AMI_matrix[j][i] = MI_score
```

```python
sns.set()

f, ax = plt.subplots(figsize = (9,6))

ax = sns.heatmap(AMI_matrix, annot=True)
ax.set_title('Adjusted Mutual information Score')
```

## Evaluate matrices

```python
def high_correlate(matrix, threshold):
    x_ind,y_ind = np.where(abs(matrix) > threshold)
    index = []
    for (x,y) in zip(x_ind,y_ind):
        if x==y: continue
        else:
            if (y,x) in index: continue
            index.append((x,y))
    return index
```

```python
cc_index = high_correlate(correlation_matrix, 0.5)
NMI_index = high_correlate(MI_matrix, 0.5)
AMI_index = high_correlate(AMI_matrix, 0.5)
```

## Classifier

1. Split dataset into 9:1 for train : test
2. Random forest classfier vs. naive bayes classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
""" Dataset split """

data_length = len(X)
test_ratio = 0.1

num_test = int(data_length * test_ratio)

train_X = X[:-num_test]
test_X = X[-num_test:]
train_Y = Y[:-num_test]
test_Y = Y[-num_test:]
```

```python
clf = GaussianNB()

# train

clf.fit(train_X, train_Y)
y_pred = clf.predict(test_X)
acc = accuracy_score(y_pred, test_Y)
```

```python
"By selecting features show change in accuracy"

np.random.seed(1010)

# k : ( 5 ~ 12 )

k_list = []
acc_list = []

for k in range(1, 13): # k is number of selected feature
    
    selector = SelectKBest(mutual_info_classif, k=k)
    X_new = selector.fit_transform(X,Y)
    
    """ Dataset split """

    data_length = len(X_new)
    test_ratio = 0.1

    num_test = int(data_length * test_ratio)

    train_X = X_new[:-num_test]
    test_X = X_new[-num_test:]
    train_Y = Y[:-num_test]
    test_Y = Y[-num_test:]
    
    clf = GaussianNB()

    # train
    clf.fit(train_X, train_Y)
    
    # predict
    y_pred = clf.predict(test_X)
    
    # evaluate
    acc = accuracy_score(y_pred, test_Y)
    
    # save
    k_list.append(k)
    acc_list.append(acc)
    
print('max accuracy is', max(acc_list))
print('when k is', k_list[np.argmax(np.array(acc_list))])
```

```python
plt.plot(k_list, acc_list, 'r-')
plt.xlabel('number of features')
plt.ylabel('accuracy')
plt.title('Naive Bayes Classifier')
plt.show()
```

```python
"By selecting features show change in accuracy"

# k : ( 5 ~ 12 )

np.random.seed(100)

k_list = []
acc_list = []

for k in range(1, 13): # k is number of selected feature
    
    selector = SelectKBest(mutual_info_classif, k=k)
    X_new = selector.fit_transform(X,Y)
    
    """ Dataset split """

    data_length = len(X_new)
    test_ratio = 0.1

    num_test = int(data_length * test_ratio)

    train_X = X_new[:-num_test]
    test_X = X_new[-num_test:]
    train_Y = Y[:-num_test]
    test_Y = Y[-num_test:]
    
    clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)

    # train
    clf.fit(train_X, train_Y)
    
    # predict
    y_pred = clf.predict(test_X)
    
    # evaluate
    acc = accuracy_score(y_pred, test_Y)
    
    # save
    k_list.append(k)
    acc_list.append(acc)
    
print('max accuracy is', max(acc_list))
print('when k is', k_list[np.argmax(np.array(acc_list))])
```

```python
plt.plot(k_list, acc_list, 'r-')
plt.xlabel('number of features')
plt.ylabel('accuracy')
plt.title('Random Forest Classifier')
plt.show()
```
